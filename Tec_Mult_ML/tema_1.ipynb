```python
import pandas as pd
from pandas.core.common import flatten

from plotnine import *
a1= np.random.normal(170,5,10)
a2= np.random.normal(170,5,100)
a3= np.random.normal(170,5,10000)
#para manipular los datos como partes de un objeto dataframe hace falta trabjar con listas y no con np arrays
#que te devuelve numpy
l_a1 = a1.tolist()
l_a2 = a2.tolist()
l_a3 = a3.tolist()
l_full = l_a1 + l_a2 + l_a3
print(len(l_full))
etiqueta = list(flatten([["n = 10"] * 10, ["n = 100"] * 100,["n = 10000"] * 10000]))
print(len(etiqueta))
d = {"Altura":l_full, "Muestra":etiqueta}
df = pd.DataFrame(d)
plot1 = (
    ggplot(df, aes(x = "Altura", color = "Muestra", 
    fill = "Muestra")) +
    geom_density(alpha = 0.1) +
    ylab("densidad de probabilidad") +
    xlab("altura (cm)") +
    theme_bw() +
    theme(legend_position = "right",
    subplots_adjust={'right': 0.8})
)
plot1.show()
```
