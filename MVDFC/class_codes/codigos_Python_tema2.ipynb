```python
import numpy as np

def VI(s, K, text):
    # Determine if it's a call or put option
    op = int(text == 'call')
    # Calculate the payoff of the option (intrinsic value)
    return np.maximum((1 - 2 * op) * (K - s), 0)

def comp(s, K, text):
    # Determine if it's a call or put option
    op = (1 - 2 * (text == 'call')) * (K - s)
    # Return 'ITM', 'OTM', or 'ATM' based on the option moneyness
    return np.where(op > 0, 'ITM', np.where(op < 0, 'OTM', 'ATM'))
```

```python
text = 'call'
s = 100
K = np.array([90, 95, 100, 105, 110])
C = np.array([12, 10, 8, 6, 4]) # call
n = len(K)
print(text)
print('\nClas|K|VI|C|VE\n')
for i in range(n):
    print(comp(s, K[i], text), '|', K[i], '|',VI(s, K[i], text), '|', C[i] - VI(s, K[i], text))
```

```python
print(text)
print(['Clas','K','VI','C','VE'])
print('\n',np.vstack((comp(s,K,text),K,VI(s,K,text),C-VI(s,K,text))).T)
```

# Opción de Compra Cubierta

# Código básico

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función
def calcular_BP_call_total(S_T, K, C):
    BP_total = S_T - K - max(S_T - K, 0) + C
    return BP_total

# Parámetros
K = 90     # Precio de ejercicio (strike price)
C = 5      # Precio de la opción

# Rango de precios S_T
S_T_values = np.linspace(0, 150, 100)

# Calcular los valores de (B/P)_{TOTAL}(S_T) para cada S_T
BP_total_values = [calcular_BP_call_total(S_T, K, C) for S_T in S_T_values]

# Crear la gráfica_put_put
plt.figure(figsize=(8, 6))
plt.plot(S_T_values, BP_total_values, label=r'$(B/P)_{\text{TOTAL}}(S_T)$', color='b')
plt.axhline(0, color='gray', lw=0.5)  # Línea horizontal en 0
plt.axvline(K, color='red', linestyle='--', label='Strike Price (K)', lw=0.8)  # Línea vertical en K
plt.title(r'Gráfico de $(B/P)_{\text{TOTAL}}(S_T)$ vs $S_T$')
plt.xlabel(r'$S_T$ (Precio del activo subyacente)')
plt.ylabel(r'$(B/P)_{\text{TOTAL}}(S_T)$')
plt.legend()
plt.grid(True)
plt.show()
```
